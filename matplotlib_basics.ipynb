```python
# --     Importing matplotlib
import matplotlib.pyplot as plt
```

```python
# plotting a siple line graph

cubes = [i**3 for i in range(10)]

# we use numbers data to plot
plt.plot(cubes)

plt.show()   # opens the matplotlibs viewer and displays plot
```

```python
# changing the label and line thickness

plt.plot(cubes,linewidth=3,color='r')
# linewidth will change the width of line and color will cahnge the color of line

plt.xlabel('Numbers', fontsize=15)
# xlabel sets label to x azis

plt.ylabel('Cubes', fontsize=15)
# y label sets label to y axis

plt.title('Cubes of Numbers',fontsize=22)
# sets title for the plot

plt.show()
```

```python
# setting background style

print(plt.style.available)
```

```python
plt.style.use('seaborn-dark-palette')
# this line must be before statring generating plot


plt.plot(cubes,linewidth=3,color='r')
plt.xlabel('Numbers', fontsize=15)
plt.ylabel('Cubes', fontsize=15)
plt.title('Cubes of Numbers',fontsize=22)
plt.show()
```

```python
# using colormap:- colormap is a series in a gradient that moves from starting to ending color.

plt.style.use('seaborn-dark-palette')
# this line must be before statring generating plot

x_val = range(10)
cubes = [x**3 for x in x_val]
plt.scatter(x_val,cubes,c=cubes,cmap=plt.cm.Accent,s=30)

plt.xlabel('Numbers', fontsize=15)
plt.ylabel('Cubes', fontsize=15)
plt.title('Cubes of Numbers',fontsize=22)
plt.show()
```

```python
# saving Plots autmatically

plt.style.use('seaborn-dark-palette')

x_val = range(1,5000)
cubes = [x**3 for x in x_val]
plt.figure(figsize=(14,10))
plt.scatter(x_val,cubes,c=cubes,cmap=plt.cm.gist_rainbow_r,s=30)

plt.xlabel('Numbers', fontsize=15)
plt.ylabel('Cubes', fontsize=15)
plt.title('Cubes of Numbers',fontsize=22)

plt.savefig('cubes_plt.png',bbox_inches='tight')
```
